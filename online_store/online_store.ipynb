##  Исходные данные

### Переданные файлы
* Игры - games.csv

### Описание данных

* **Name** — название игры
* **Platform** — платформа
* **Year_of_Release** — год выпуска
* **Genre** — жанр игры
* **NA_sales** — продажи в Северной Америке (миллионы проданных копий)
* **EU_sales** — продажи в Европе (миллионы проданных копий)
* **JP_sales** — продажи в Японии (миллионы проданных копий)
* **Other_sales** — продажи в других странах (миллионы проданных копий)
* **Critic_Score** — оценка критиков (максимум 100)
* **User_Score** — оценка пользователей (максимум 10)
* **Rating** — рейтинг от организации ESRB

### Описание возрастных категорий

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Рейтинги ESRB:
* «EC» («Early childhood») — «Для детей младшего возраста»
* «E» («Everyone») — «Для всех»
* «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»
* «T» («Teen») — «Подросткам»
* «M» («Mature») — «Для взрослых»
* «AO» («Adults Only 18+») — «Только для взрослых»
* «RP» («Rating Pending») — «Рейтинг ожидается»

## Предобработка данных

### Изменение типа данных

Приведем названия столбцов к нижнему регистру.

```python
games_df.columns = games_df.columns.str.lower()
```

Приведем столбец **year_of_release** к целочисленному типу *int*.

```python
games_df['year_of_release'] = games_df['year_of_release'].astype('Int64')
```

Узнаем, почему столбец **user_score** имеет строковый тип, а не вещественный.

```python
games_df['user_score'].unique()
```

Среди прочих оценок, есть значение **tbd**, что означает "To Be Determined" - будет определено. Раз оценки не определены, присвоим им значение **NaN**, а столбец приведем к вещественному типу **float**.

```python
games_df['user_score'] = games_df['user_score'].replace('tbd', np.nan, regex=True)
games_df['user_score'] = games_df['user_score'].astype('float64')
```

**Вывод**

На данном этапе предобработки данных:
* названия столбцов приведены к нижнему регистру.
* столбец **year_of_release** приведен к типу *int64*.
* в столбеце **user_score** значения **tbd** заменены на **nan**
* столбец **user_score** приведен к типу *float*

```python
print(games_df['year_of_release'].isnull().sum())
games_df = games_df.dropna(subset=['year_of_release']).reset_index(drop=True)
print(games_df['year_of_release'].isnull().sum())
```

```python
# делаем срез данных по названию платформы
df_ps3 = popular_platforms_df.query('platform == "PS3" and critic_score != "unknown" and user_score != "unknown"')

# размечаем пространство для вывода графиков
fig, (ax1, ax2) = plt.subplots(
    nrows=1, ncols=2,
    figsize=(10, 5)
)

# построение и оформление графика рассеивания для отзывов критиков
ax1.scatter(x=df_ps3.critic_score,
            y=df_ps3.total_sales,
            alpha=0.35,
            color='black');
ax1.set_title('PS3: critic_score VS total_sales')
ax1.set_xlabel('оценки критиков')
ax1.set_ylabel('общие продажи')

# построение и оформление графика рассеивания для отзывов пользователей
ax2.scatter(x=df_ps3.user_score,
            y=df_ps3.total_sales,
            alpha=0.35,
            color='black');
ax2.set_title('PS3: user_score VS total_sales')
ax2.set_xlabel('оценки игроков')
ax2.set_ylabel('общие продажи');
```

```python
# Подсчет и вывод на экран коэффициентов корреляции
print('PS3:')
df_ps3['critic_score'] = pd.to_numeric(df_ps3['critic_score'])
print('Коэффициент корреляции глобальных продаж и оценки критиков:', round(df_ps3['total_sales'].corr(df_ps3['critic_score']), 3))
df_ps3['user_score'] = pd.to_numeric(df_ps3['user_score'])
print('Коэффициент корреляции глобальных продаж и оценки пользователей:', round(df_ps3['total_sales'].corr(df_ps3['user_score']), 3))
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC

$H_0$ - cредние пользовательские рейтинги платформ Xbox One и PC одинаковые

$H_1$ - cредние пользовательские рейтинги платформ Xbox One и PC разные

```python
period2_df = period_df.dropna()
#period2_df['user_score'] = pd.to_numeric(period2_df['user_score'])
xone = period2_df.query('platform == "XOne"')['user_score']
pc = period2_df.query('platform == "PC" ')['user_score']

alpha = .05

results = st.ttest_ind(
    xone, 
    pc, 
    equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
xone_pc = period2_df.query('platform == "XOne" or platform == "PC"')
xone_pc.pivot_table(index='platform', values='user_score', aggfunc='mean')
```

**Вывод**

С большой долей вероятности средние пользовательские рейтинги платформ XOne и PC не имеют статистически значимых отличий

### Средние пользовательские рейтинги жанров Action и Sports

$H_0$ - cредние пользовательские рейтинги жанров Action и Sports одинаковые

$H_1$ - cредние пользовательские рейтинги жанров Action и Sports разные

```python
action = period2_df.query('genre == "Action"')['user_score']
sports = period2_df.query('genre == "Sports"')['user_score']

alpha = .05

results = st.ttest_ind(
    action, 
    sports, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**

Разница между жанрами не большая, но ощутимая.

Гипотезу о равенстве среднего пользовательского рейтинга для XOne и PC не получилось отвергнуть.
